# Maps with Markers

```python
df_indicents = pd.read_csv("https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv")

print("Dataset downloaded and read into a pandas dataframe!")
```

```python
df_indicents.head()
```

```python
df_indicents.shape
```

```python
limit = 100
df_indicents = df_indicents.iloc[0: limit, :]
```

```python
df_indicents.shape
```

```python
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

sanfran_map
```

```python
indicents = folium.map.FeatureGroup()

for lat, lng, in zip(df_indicents.Y, df_indicents.X):
    indicents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
sanfran_map.add_child(indicents)
```

```python
indicents = folium.map.FeatureGroup()

for lat, lng, in zip(df_indicents.Y, df_indicents.X):
    indicents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
latitudes = list(df_indicents.Y)
longitudes = list(df_indicents.X)
labels = list(df_indicents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)
    
sanfran_map.add_child(indicents)
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, lng, label in zip(df_indicents.Y, df_indicents.X, df_indicents.Category):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)
    
sanfran_map
```

```python
from folium import plugins

sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

indicents = plugins.MarkerCluster().add_to(sanfran_map)

for lat, lng, label in zip(df_indicents.Y, df_indicents.X, df_indicents.Category):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(indicents)
    
sanfran_map
```

```python
df_can = pd.read_excel("https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
                      sheet_name="Canada by Citizenship",
                      skiprows=range(20),
                      skipfooter=2)

print("Data downloaded and read into a dataframe!")
```

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_can.rename(columns={'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'}, inplace=True)

df_can.columns = list(map(str, df_can.columns))

df_can['Total'] = df_can.sum(axis=1)

years = list(map(str, range(1980, 2014)))
print('data dimensions:', df_can.shape)
```

```python
!wget --quiet https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/world_countries.json

print("GeoJSON file downloaded")
```

```python
world_geo = r'world_countries.json'

world_map = folium.Map(location=[0, 0], zoom_start=2, tiles='Mapbox Bright')

world_map.choropleth(
    geo_data=world_geo,
    data=df_can,
    columns=['Country', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Immigration to Canada'
)

world_map
```
